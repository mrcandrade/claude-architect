```python
import os
from dotenv import load_dotenv
import anthropic

load_dotenv()
client = anthropic.Anthropic(api_key=os.getenv('ANTHROPIC_API_KEY'))
HAIKU  = 'claude-haiku-4-5-20251001'
SONNET = 'claude-sonnet-4-5'

def ask(prompt, system=None, model=HAIKU, max_tokens=1024):
    kw = dict(model=model, max_tokens=max_tokens,
              messages=[{'role':'user','content':prompt}])
    if system: kw['system'] = system
    return client.messages.create(**kw).content[0].text

print('Pronto!')
```

## Viabilidade Tecnologica com Claude

Vamos usar Claude como consultor tecnico para avaliar a viabilidade de uma
tecnologia especifica. O segredo esta no **system prompt** que define o papel
e a expertise do modelo.

```python
system_engenheiro = """Voce e um engenheiro consultor senior com 20 anos de experiencia
em sensores industriais, sistemas embarcados e manutencao preditiva.
Responda de forma estruturada com: VIABILIDADE (Alta/Media/Baixa),
JUSTIFICATIVA, RISCOS TECNICOS, ALTERNATIVAS e RECOMENDACAO FINAL.
Use dados tecnicos reais quando possivel."""

pergunta_viabilidade = """Estamos avaliando o uso de sensores piezoeletricos de vibracao
para manutencao preditiva em motores eletricos industriais (5-50 HP).

Restricoes do projeto:
- Custo maximo por sensor: R$ 150
- Precisao minima: deteccao de desbalanceamento > 0.5mm/s RMS
- Tamanho maximo: 30x30x15mm
- Temperatura de operacao: -10C a 85C
- Comunicacao: wireless (LoRa ou BLE)
- Autonomia de bateria: minimo 2 anos com leitura a cada 15 min

Avalie a viabilidade tecnica dessa solucao."""

resp = ask(pergunta_viabilidade, system=system_engenheiro, model=SONNET, max_tokens=2048)
print(resp)
```

### Refinando a Analise

Podemos pedir mais detalhes sobre pontos especificos da viabilidade.

```python
refinamento = """Com base em sensores MEMS acelerometros (tipo ADXL345 ou LIS3DH),
e possivel atingir a resolucao de 0.5mm/s RMS para vibracao em motores?
Compare com sensores piezoeletricos tradicionais (tipo PCB Piezotronics 603C01)
em termos de: custo, sensibilidade, faixa de frequencia e durabilidade.
Formate como tabela comparativa."""

resp2 = ask(refinamento, system=system_engenheiro, model=SONNET, max_tokens=2048)
print(resp2)
```

## Analise de Restricoes Fisicas

Claude pode ajudar a avaliar limitacoes fisicas de uma solucao proposta,
identificando gargalos antes de investir em prototipagem.

```python
system_fisico = """Voce e um fisico especialista em materiais e energia solar.
Responda com rigor tecnico, citando limites teoricos conhecidos
(como limite de Shockley-Queisser) e dados de pesquisas recentes.
Estruture: ANALISE TEORICA, LIMITACOES PRATICAS, ESTADO DA ARTE, CONCLUSAO."""

pergunta_fisica = """Estamos projetando um dispositivo IoT com celula solar de filme fino
para alimentacao autonoma. Especificacoes:

- Material: Perovskita (CH3NH3PbI3)
- Area disponivel: 5cm x 5cm (25 cm2)
- Eficiencia alvo: 20% em condicoes reais (nao laboratorio)
- Vida util desejada: 5 anos em ambiente externo
- Potencia necessaria: 500mW medio ao longo do dia
- Local: regiao subtropical (irradiancia media 4.5 kWh/m2/dia)

Quais sao as restricoes fisicas fundamentais dessa proposta?
E viavel com a tecnologia atual?"""

resp3 = ask(pergunta_fisica, system=system_fisico, model=SONNET, max_tokens=2048)
print(resp3)
```

```python
# Pergunta de follow-up sobre alternativas
alternativas = """Dado que perovskita ainda tem problemas de durabilidade,
quais alternativas de filme fino seriam mais adequadas para 5 anos
em ambiente externo? Compare CIGS, CdTe e silicio amorfo (a-Si)
para a mesma aplicacao (25cm2, 500mW medio).
Inclua custo estimado por watt e eficiencia tipica."""

resp4 = ask(alternativas, system=system_fisico, model=HAIKU)
print(resp4)
```

## Analise Economica Preliminar

Claude pode ajudar a estruturar uma analise de custos, gerando tabelas
e estimativas iniciais para tomada de decisao.

```python
system_economico = """Voce e um analista de custos e viabilidade economica
para projetos de hardware eletronico. Gere estimativas realistas
baseadas em precos tipicos de componentes (Digikey, Mouser, LCSC)
e custos de engenharia no Brasil. Use formato estruturado com valores em BRL."""

pergunta_bom = """Gere uma estimativa de BOM (Bill of Materials) para um sensor
de vibracao wireless para manutencao preditiva com estas specs:

- MCU: ESP32-C3 ou nRF52832
- Acelerometro MEMS: 3 eixos, 16-bit
- Comunicacao: BLE 5.0
- Bateria: Li-SOCl2 3.6V (longa duracao)
- PCB: 2 camadas, 25x25mm
- Encapsulamento: IP67

Retorne como dicionario Python com a seguinte estrutura:
bom = {
    'componente': [...],
    'descricao': [...],
    'qtd': [...],
    'preco_unit_brl': [...],
    'subtotal_brl': [...]
}

Inclua tambem custos de montagem SMT para lote de 100 unidades."""

resp5 = ask(pergunta_bom, system=system_economico, model=SONNET, max_tokens=2048)
print(resp5)
```

```python
# Analise de custos de desenvolvimento
custos_dev = """Estime os custos de desenvolvimento (NRE - Non-Recurring Engineering)
para o projeto do sensor de vibracao wireless descrito acima.

Considere:
- Design de hardware (esquematico + PCB layout)
- Firmware (BLE stack + aquisicao + processamento FFT)
- App mobile basico para configuracao
- Backend/dashboard simples para visualizacao
- Certificacoes (Anatel, CE basico)
- Testes e validacao

Retorne como tabela com: etapa, horas estimadas, custo/hora medio, total.
Compare o custo total de desenvolvimento com o preco de mercado de solucoes
similares (ex: SKF, Fluke, Banner Engineering)."""

resp6 = ask(custos_dev, system=system_economico, model=SONNET, max_tokens=2048)
print(resp6)
```

## Pesquisa de Mercado Assistida

Claude pode analisar descricoes de mercado e identificar concorrentes,
diferenciais competitivos e riscos de negocio.

```python
system_mercado = """Voce e um analista de mercado especializado em IoT industrial
e manutencao preditiva. Analise de forma objetiva, identificando
oportunidades e ameacas reais. Use framework: CONCORRENTES,
DIFERENCIAIS POSSIVEIS, RISCOS DE MERCADO, OPORTUNIDADES, RECOMENDACAO."""

descricao_mercado = """Nosso produto e um sensor de vibracao wireless de baixo custo
(alvo: R$ 300-500 por unidade) para manutencao preditiva em
pequenas e medias industrias no Brasil.

Publico alvo: fabricas com 10-100 motores eletricos que hoje
fazem manutencao corretiva ou preventiva baseada em calendario.

Proposta de valor: sensor plug-and-play com dashboard web,
sem necessidade de engenheiro de confiabilidade dedicado.
Algoritmo de IA embarcado para deteccao automatica de anomalias.

Modelo de negocio: hardware + assinatura mensal (SaaS) para
dashboard e alertas. Preco alvo: R$ 50/sensor/mes.

Analise este mercado e identifique concorrentes, diferenciais e riscos."""

resp7 = ask(descricao_mercado, system=system_mercado, model=SONNET, max_tokens=2048)
print(resp7)
```

```python
# Analise SWOT estruturada
swot = """Com base na analise anterior, gere uma matriz SWOT detalhada
para o projeto de sensor de vibracao wireless no mercado brasileiro.

Formate como dicionario Python:
swot = {
    'forcas': ['...', '...'],
    'fraquezas': ['...', '...'],
    'oportunidades': ['...', '...'],
    'ameacas': ['...', '...']
}

Inclua pelo menos 5 itens em cada categoria."""

resp8 = ask(swot, system=system_mercado, model=HAIKU)
print(resp8)
```

## Gerando Relatorio de Viabilidade

Agora vamos combinar todas as analises anteriores em um relatorio
estruturado de viabilidade, usando Claude para sintetizar e organizar.

```python
system_relatorio = """Voce e um gerente de projetos tecnico que cria relatorios
de viabilidade executivos. O relatorio deve ser claro, objetivo
e conter recomendacoes acionaveis. Use Markdown para formatacao."""

prompt_relatorio = """Gere um RELATORIO DE VIABILIDADE EXECUTIVO para o seguinte projeto:

PROJETO: Sensor de Vibracao Wireless para Manutencao Preditiva
EMPRESA: EnfiTec Servicos Ltda.

O relatorio deve conter estas secoes:

1. RESUMO EXECUTIVO (max 5 linhas)
2. VIABILIDADE TECNICA
   - Tecnologia de sensoriamento: acelerometro MEMS 3-eixos
   - Comunicacao: BLE 5.0
   - Processamento: ESP32-C3 com FFT embarcada
   - Autonomia alvo: 2 anos com bateria Li-SOCl2
3. VIABILIDADE ECONOMICA
   - BOM estimado: ~R$ 85/unidade (lote 100)
   - Preco de venda alvo: R$ 350-500
   - Margem bruta estimada: 65-75%
   - NRE estimado: R$ 150-200k
   - Break-even: ~400 unidades
4. ANALISE DE MERCADO
   - TAM/SAM/SOM estimados para Brasil
   - Principais concorrentes
   - Diferencial competitivo
5. RISCOS E MITIGACOES (top 5)
6. CRONOGRAMA SUGERIDO (fases e marcos)
7. RECOMENDACAO FINAL (Go/No-Go com condicoes)

Use dados realistas e coerentes entre si."""

relatorio = ask(prompt_relatorio, system=system_relatorio, model=SONNET, max_tokens=4096)
print(relatorio)
```

```python
# Salvar o relatorio em arquivo Markdown
with open('relatorio_viabilidade_sensor.md', 'w', encoding='utf-8') as f:
    f.write(relatorio)

print('Relatorio salvo em: relatorio_viabilidade_sensor.md')
print(f'Tamanho: {len(relatorio)} caracteres')
```
