# 02 - Latencia vs Qualidade
> Trade-offs e roteamento inteligente entre modelos

```python
import os
from dotenv import load_dotenv
import anthropic

load_dotenv()
client = anthropic.Anthropic(api_key=os.getenv('ANTHROPIC_API_KEY'))
HAIKU  = 'claude-haiku-4-5-20251001'
SONNET = 'claude-sonnet-4-5'

def ask(prompt, system=None, model=HAIKU, max_tokens=1024):
    kw = dict(model=model, max_tokens=max_tokens,
              messages=[{'role':'user','content':prompt}])
    if system: kw['system'] = system
    return client.messages.create(**kw).content[0].text

print('Pronto!')
```

```python
import time

def complexidade(q):
    r=ask(f'Classifique a complexidade: "{q}" - responda: simples, media ou complexa',model=HAIKU)
    return r.strip().lower()

def rotear(q):
    c=complexidade(q)
    m={'simples':HAIKU,'media':HAIKU,'complexa':SONNET}.get(c,HAIKU)
    return m,c

qs=['Qual a capital do Brasil?',
    'Compare REST e GraphQL para e-commerce.',
    'Projete arquitetura de recomendacao em tempo real para 10M usuarios.']

for q in qs:
    m,c=rotear(q)
    print(f'[{c:8}] -> {m.split("-")[1]:7} | {q[:55]}')
```
