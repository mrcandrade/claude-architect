# 05 - Dashboard de Qualidade

```python
import os
from dotenv import load_dotenv
import anthropic

load_dotenv()
client = anthropic.Anthropic(api_key=os.getenv('ANTHROPIC_API_KEY'))
HAIKU  = 'claude-haiku-4-5-20251001'
SONNET = 'claude-sonnet-4-5'

def ask(prompt, system=None, model=HAIKU, max_tokens=1024):
    kw = dict(model=model, max_tokens=max_tokens,
              messages=[{'role':'user','content':prompt}])
    if system: kw['system'] = system
    return client.messages.create(**kw).content[0].text

print('Pronto!')
```

```python
import matplotlib.pyplot as plt, matplotlib.dates as mdates
import datetime, random

hist=[]
for i in range(30):
    data=datetime.datetime.now()-datetime.timedelta(days=30-i)
    hist.append({'data':data,'acc':min(0.98,0.72+i*0.008+random.gauss(0,0.02)),
                 'lat':round(random.gauss(1.2,0.2),2)})

fig,(ax1,ax2)=plt.subplots(1,2,figsize=(12,4))
fig.suptitle('Dashboard de Qualidade - Claude',fontsize=13,fontweight='bold')

datas=[h['data']for h in hist]; accs=[h['acc']for h in hist]; lats=[h['lat']for h in hist]

ax1.plot(datas,accs,'b-o',ms=4)
ax1.axhline(0.8,color='r',linestyle='--',label='Target 80%')
ax1.fill_between(datas,accs,0.8,where=[a>=0.8 for a in accs],alpha=0.2,color='green')
ax1.set_title('Acuracia ao longo do tempo'); ax1.legend()
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%d/%m'))

ax2.bar(range(len(lats)),lats,color='steelblue',alpha=0.7)
ax2.axhline(sum(lats)/len(lats),color='r',linestyle='--',label=f'Media')
ax2.set_title('Latencia'); ax2.legend()

plt.tight_layout()
plt.savefig('dashboard.png',dpi=120,bbox_inches='tight')
plt.show(); print('dashboard.png salvo')
```
