```python
import os
from dotenv import load_dotenv
import anthropic

load_dotenv()
client = anthropic.Anthropic(api_key=os.getenv('ANTHROPIC_API_KEY'))
HAIKU  = 'claude-haiku-4-5-20251001'
SONNET = 'claude-sonnet-4-5'

def ask(prompt, system=None, model=HAIKU, max_tokens=1024):
    kw = dict(model=model, max_tokens=max_tokens,
              messages=[{'role':'user','content':prompt}])
    if system: kw['system'] = system
    return client.messages.create(**kw).content[0].text

print('Pronto!')
```

```python
import re, json

def exact(saida,esp): return {'passou':saida.strip().lower()==esp.strip().lower(),'tipo':'exact'}
def contem(saida,esp): return {'passou':esp.lower() in saida.lower(),'tipo':'contains'}
def regex_match(saida,pad): return {'passou':bool(re.search(pad,saida,re.I)),'tipo':'regex'}
def json_valido(saida,keys=None):
    try:
        d=json.loads(saida)
        return {'passou':all(k in d for k in (keys or [])),'tipo':'json'}
    except: return {'passou':False,'tipo':'json'}
def comprimento(saida,mn=10,mx=500):
    return {'passou':mn<=len(saida)<=mx,'tipo':'len'}

s='{"sentimento":"positivo","score":0.9}'
for fn,args in[(contem,(s,'positivo')),(json_valido,(s,['sentimento','score'])),(comprimento,(s,20,100))]:
    r=fn(*args); print(f"{'OK' if r['passou'] else 'FALHOU'} [{r['tipo']}]")
```
