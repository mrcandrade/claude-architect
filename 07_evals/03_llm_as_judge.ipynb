```python
import os
from dotenv import load_dotenv
import anthropic

load_dotenv()
client = anthropic.Anthropic(api_key=os.getenv('ANTHROPIC_API_KEY'))
HAIKU  = 'claude-haiku-4-5-20251001'
SONNET = 'claude-sonnet-4-5'

def ask(prompt, system=None, model=HAIKU, max_tokens=1024):
    kw = dict(model=model, max_tokens=max_tokens,
              messages=[{'role':'user','content':prompt}])
    if system: kw['system'] = system
    return client.messages.create(**kw).content[0].text

print('Pronto!')
```

```python
import json

def julgar(pergunta, resposta, criterios=None):
    crit=criterios or ['precisao','clareza','completude']
    prompt=f'''
    Avalie a resposta abaixo nos criterios: {", ".join(crit)}
    Pergunta: {pergunta}\nResposta: {resposta}
    JSON: {{"score_geral":0-10,"passou":bool,"justificativa":"..."}}
    '''
    try: return json.loads(ask(prompt,model=HAIKU))
    except: return {'score_geral':0,'passou':False,'justificativa':'Erro'}

q='Explique recursao em programacao.'
respostas=[
    'Recursao e quando uma funcao chama a si mesma para resolver subproblemas menores.',
    'e tipo quando a coisa chama ela mesma sabe',
    ask(q,model=HAIKU)
]
nomes=['Boa','Ruim','Claude']
for nome,r in zip(nomes,respostas):
    j=julgar(q,r)
    print(f"{'OK' if j['passou'] else 'FALHOU'} {nome}: {j['score_geral']}/10 - {j['justificativa'][:70]}")
```
