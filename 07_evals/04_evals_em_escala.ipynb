```python
import os
from dotenv import load_dotenv
import anthropic

load_dotenv()
client = anthropic.Anthropic(api_key=os.getenv('ANTHROPIC_API_KEY'))
HAIKU  = 'claude-haiku-4-5-20251001'
SONNET = 'claude-sonnet-4-5'

def ask(prompt, system=None, model=HAIKU, max_tokens=1024):
    kw = dict(model=model, max_tokens=max_tokens,
              messages=[{'role':'user','content':prompt}])
    if system: kw['system'] = system
    return client.messages.create(**kw).content[0].text

print('Pronto!')
```

```python
import time
from concurrent.futures import ThreadPoolExecutor

DATASET=[{'id':f's{i}','input':inp,'label':lbl}
    for i,(inp,lbl) in enumerate([
        ('Adorei, produto incrivel!','positivo'),
        ('Chegou quebrado, horrivel.','negativo'),
        ('Entrega no prazo.','neutro'),
        ('Superou minhas expectativas!','positivo'),
        ('Nao vale o preco.','negativo'),
    ])]

SYS='Classifique como positivo, negativo ou neutro. Responda uma palavra.'

def rodar(caso):
    t0=time.time()
    saida=ask(caso['input'],system=SYS,model=HAIKU).strip().lower()
    return {**caso,'saida':saida,'passou':caso['label']in saida,'dt':round(time.time()-t0,2)}

with ThreadPoolExecutor(max_workers=3) as ex:
    res=list(ex.map(rodar,DATASET))

acuracia=sum(r['passou']for r in res)/len(res)
print(f'Acuracia: {acuracia:.0%}')
for r in res:
    print(f"  {'OK' if r['passou'] else 'FALHOU'} {r['input'][:30]:30} -> {r['saida']:10} ({r['label']})")
```
