```python
import os
from dotenv import load_dotenv
import anthropic

load_dotenv()
client = anthropic.Anthropic(api_key=os.getenv('ANTHROPIC_API_KEY'))
HAIKU  = 'claude-haiku-4-5-20251001'
SONNET = 'claude-sonnet-4-5'

def ask(prompt, system=None, model=HAIKU, max_tokens=1024):
    kw = dict(model=model, max_tokens=max_tokens,
              messages=[{'role':'user','content':prompt}])
    if system: kw['system'] = system
    return client.messages.create(**kw).content[0].text

print('Pronto!')
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

model=SentenceTransformer('all-MiniLM-L6-v2')

def sim(a,b): return float(np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b)))

frases=['O gato sentou no tapete',
        'Um felino descansou no tapete',
        'O cachorro correu no parque',
        'Python e uma linguagem de programacao']

embs=model.encode(frases)
ref=embs[0]
print(f'Referencia: "{frases[0]}"\n')
for f,e in zip(frases[1:],embs[1:]):
    s=sim(ref,e)
    print(f'{s:.3f} {"#"*int(s*20)} "{f}"')
```
