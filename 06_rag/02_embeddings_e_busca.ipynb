```python
import os
from dotenv import load_dotenv
import anthropic

load_dotenv()
client = anthropic.Anthropic(api_key=os.getenv('ANTHROPIC_API_KEY'))
HAIKU  = 'claude-haiku-4-5-20251001'
SONNET = 'claude-sonnet-4-5'

def ask(prompt, system=None, model=HAIKU, max_tokens=1024):
    kw = dict(model=model, max_tokens=max_tokens,
              messages=[{'role':'user','content':prompt}])
    if system: kw['system'] = system
    return client.messages.create(**kw).content[0].text

print('Pronto!')
```

```python
import chromadb
from sentence_transformers import SentenceTransformer

model=SentenceTransformer('all-MiniLM-L6-v2')
col=chromadb.Client().create_collection('docs')

docs=['Python foi criado em 1991 por Guido van Rossum.',
      'FastAPI e um framework web moderno para Python.',
      'Docker containeriza aplicacoes e suas dependencias.',
      'RAG combina busca vetorial com geracao por LLMs.',
      'O Claude e desenvolvido pela Anthropic.']

col.upsert(ids=[f'd{i}'for i in range(len(docs))],
           embeddings=model.encode(docs).tolist(), documents=docs)

def buscar(q,n=2):
    res=col.query(query_embeddings=model.encode([q]).tolist(),n_results=n)
    return list(zip(res['documents'][0],res['distances'][0]))

for p in ['Como criar uma API em Python?','O que e containerizacao?']:
    print(f'\nBusca: "{p}"')
    for doc,dist in buscar(p): print(f'  [{1-dist:.2f}] {doc[:70]}')
```
