## Configuracao Inicial

Importamos o cliente Anthropic e definimos os modelos e a funcao auxiliar `ask()`.

```python
from anthropic import Anthropic

client = Anthropic()

HAIKU  = "claude-haiku-4-20250414"
SONNET = "claude-sonnet-4-20250514"


def ask(prompt: str, *, model: str = SONNET, system: str = "", temperature: float = 0.3) -> str:
    """Envia um prompt ao Claude e retorna a resposta como texto."""
    msgs = [{"role": "user", "content": prompt}]
    kw = {"model": model, "max_tokens": 4096, "messages": msgs}
    if system:
        kw["system"] = system
    if temperature is not None:
        kw["temperature"] = temperature
    resp = client.messages.create(**kw)
    return resp.content[0].text


print("Cliente Anthropic pronto.")
```

## Estruturando um Relatorio de Laboratorio

Relatorios de laboratorio seguem uma estrutura padrao, mas alunos frequentemente
esquecem secoes ou organizam mal o conteudo. Vamos pedir ao Claude para gerar
um template e depois preencher secoes a partir de dados brutos.

```python
# Pedir um template de relatorio para experimento de optica
template = ask(
    "Gere a estrutura completa (template) de um relatorio de laboratorio de Fisica "
    "para um experimento de optica (difracao de fenda simples). "
    "Inclua todas as secoes padrao (titulo, resumo, introducao, fundamentacao teorica, "
    "materiais e metodos, resultados, discussao, conclusao, referencias). "
    "Para cada secao, escreva 1-2 frases explicando o que deve conter. "
    "Use formato Markdown. Escreva em portugues sem acentos.",
    system="Voce e um professor de Fisica Experimental da UFRGS. Responda em portugues sem acentos."
)
print(template)
```

```python
# Agora, preencher a secao de resultados a partir de dados brutos
dados_brutos = """
Experimento: Difracao de fenda simples
Largura da fenda: a = 0.1 mm
Comprimento de onda do laser: lambda = 632.8 nm (HeNe)
Distancia fenda-anteparo: L = 1.50 m
Posicoes dos minimos de difracao medidos (em mm a partir do centro):
  m=1: +9.6, -9.4
  m=2: +19.1, -19.3
  m=3: +28.5, -28.8
Incerteza na posicao: +/- 0.3 mm
Incerteza em L: +/- 0.005 m
"""

resultados = ask(
    f"A partir dos seguintes dados experimentais brutos, escreva a secao de Resultados "
    f"de um relatorio de laboratorio. Inclua:\n"
    f"- Tabela organizada com os dados\n"
    f"- Calculo da largura da fenda experimental usando a = m*lambda*L / y_m\n"
    f"- Propagacao de incertezas\n"
    f"- Comparacao com o valor nominal\n\n"
    f"Dados:\n{dados_brutos}\n"
    f"Escreva em portugues sem acentos, formato academico.",
    system="Voce e um estudante de Engenharia Fisica escrevendo um relatorio de lab. "
           "Use notacao cientifica quando apropriado. Escreva sem acentos."
)
print(resultados)
```

## Revisao e Melhoria de Texto Tecnico

Uma das aplicacoes mais uteis do Claude e revisar textos tecnicos, melhorando
clareza, corrigindo erros de logica e mantendo o rigor cientifico.

```python
# Paragrafo mal escrito para revisao
texto_ruim = """
No experimento foi usado um laser que emite luz e essa luz passa pela fenda e
depois a gente ve o padrao na parede. O padrao tem partes claras e escuras que
sao causadas por causa da difracao que acontece quando a luz passa pela fenda.
Os resultados foram mais ou menos bons e deu perto do valor que era pra dar.
A fisica da difracao e bem importante e tem muitas aplicacoes tipo em
espectroscopia e essas coisas.
"""

revisao = ask(
    f"Revise o paragrafo abaixo de um relatorio de Fisica. O texto esta mal escrito "
    f"e informal. Melhore a clareza, use linguagem academica formal, mantenha o "
    f"rigor tecnico e adicione termos cientificos apropriados.\n\n"
    f"TEXTO ORIGINAL:\n{texto_ruim}\n\n"
    f"Forneca:\n"
    f"1. O texto revisado\n"
    f"2. Uma lista das principais mudancas feitas e por que\n"
    f"Escreva em portugues sem acentos.",
    system="Voce e um revisor academico especializado em textos de Fisica. "
           "Mantenha precisao cientifica. Escreva sem acentos."
)
print("=" * 60)
print("TEXTO ORIGINAL (problematico):")
print("=" * 60)
print(texto_ruim)
print("=" * 60)
print("TEXTO REVISADO + COMENTARIOS:")
print("=" * 60)
print(revisao)
```

## Gerando Secoes de Metodologia

A secao de metodologia precisa ser clara o suficiente para que outro pesquisador
reproduza o experimento. Vamos transformar anotacoes de caderno de laboratorio
em uma secao de metodologia formal.

```python
# Notas de caderno de laboratorio (formato bullet-point)
notas_lab = """
- montei o trilho optico na mesa com antivibracoes
- laser HeNe 632.8nm, potencia 5mW
- coloquei a fenda no suporte, ajustei pra 0.1mm com o micrometro
- anteparo a 1.5m da fenda (medi com trena)
- papel milimetrado no anteparo pra medir posicoes
- tirei foto do padrao com celular pra analisar depois
- marquei os minimos no papel com lapis
- medi as posicoes dos minimos com regua milimetrada
- repeti 3 vezes cada medida
- temperatura da sala: 22 graus C
- sala escura (apaguei as luzes)
"""

metodologia = ask(
    f"Transforme as seguintes anotacoes de caderno de laboratorio em uma secao de "
    f"Materiais e Metodos formal, em formato academico. A secao deve:\n"
    f"- Ser escrita na terceira pessoa do singular ou voz passiva\n"
    f"- Listar equipamentos com especificacoes\n"
    f"- Descrever o procedimento de forma reproduzivel\n"
    f"- Mencionar condicoes experimentais relevantes\n"
    f"- Separar em subsecoes: Materiais e Procedimento Experimental\n\n"
    f"NOTAS DO CADERNO:\n{notas_lab}\n"
    f"Escreva em portugues sem acentos, estilo academico.",
    system="Voce e um pesquisador em Fisica Experimental. "
           "Escreva secoes de metodologia claras e reproduziveis. Sem acentos."
)
print(metodologia)
```

## Escrevendo Introducoes e Conclusoes

Introducoes devem contextualizar o problema e motivar o trabalho.
Conclusoes devem resumir achados sem exagerar a importancia dos resultados.

```python
# Gerar uma introducao para o experimento de difracao
introducao = ask(
    "Escreva a secao de Introducao (3-4 paragrafos) para um relatorio de laboratorio "
    "sobre difracao de fenda simples. A introducao deve:\n"
    "1. Contextualizar o fenomeno da difracao na Fisica ondulatoria\n"
    "2. Apresentar brevemente o modelo teorico (principio de Huygens-Fresnel)\n"
    "3. Mencionar aplicacoes praticas da difracao\n"
    "4. Declarar o objetivo do experimento\n\n"
    "Nivel: graduacao em Engenharia Fisica. Escreva em portugues sem acentos.",
    system="Voce e um estudante de Engenharia Fisica da UFRGS. "
           "Escreva de forma clara e academica. Sem acentos."
)
print("INTRODUCAO GERADA:")
print("-" * 40)
print(introducao)
```

```python
# Gerar uma conclusao adequada (sem exageros)
conclusao = ask(
    "Escreva a secao de Conclusao (2-3 paragrafos) para o relatorio de difracao. "
    "Considere que:\n"
    "- O valor experimental da largura da fenda foi a = (0.099 +/- 0.003) mm\n"
    "- O valor nominal era a = 0.100 mm\n"
    "- O erro relativo foi menor que 1%\n"
    "- As principais fontes de incerteza foram a medida das posicoes dos minimos\n\n"
    "IMPORTANTE: A conclusao deve:\n"
    "- Resumir os achados sem repetir a secao de resultados\n"
    "- NAO exagerar a importancia (nao dizer 'resultados extraordinarios')\n"
    "- Discutir limitacoes do metodo\n"
    "- Sugerir melhorias para trabalhos futuros\n"
    "Escreva em portugues sem acentos.",
    system="Voce e um estudante de Engenharia Fisica. Seja honesto sobre limitacoes. "
           "Evite hiperboles. Escreva sem acentos."
)
print("CONCLUSAO GERADA:")
print("-" * 40)
print(conclusao)
```

## Traducao Tecnica Portugues-Ingles

Publicar em periodicos internacionais exige traducao precisa do portugues para
o ingles. Termos tecnicos devem ser traduzidos corretamente -- erros comuns
incluem falsos cognatos e traducoes literais de expressoes idiomaticas.

```python
# Texto tecnico em portugues para traduzir
texto_pt = """
Neste trabalho, investigou-se o comportamento da difracao de Fraunhofer
em uma fenda simples. O arranjo experimental consistiu em um laser de
helio-neonio operando no comprimento de onda de 632,8 nm, uma fenda
ajustavel com micrometro e um anteparo posicionado a 1,50 m da fenda.
Os padroes de difracao foram registrados e as posicoes dos minimos de
intensidade foram medidas com incerteza de 0,3 mm. A largura da fenda
foi determinada experimentalmente como (0,099 +/- 0,003) mm, em
concordancia com o valor nominal de 0,100 mm dentro da faixa de
incerteza experimental.
"""

traducao = ask(
    f"Traduza o seguinte texto tecnico de Fisica do portugues para o ingles. \n\n"
    f"TEXTO ORIGINAL:\n{texto_pt}\n\n"
    f"Apos a traducao, forneca:\n"
    f"1. O texto traduzido\n"
    f"2. Uma tabela com os termos tecnicos chave e suas traducoes\n"
    f"3. Armadilhas comuns de traducao neste contexto (falsos cognatos, "
    f"traducoes literais a evitar)\n\n"
    f"Exemplo de armadilhas:\n"
    f"- 'realizar um experimento' -> 'to perform/conduct an experiment' "
    f"(NAO 'to realize')\n"
    f"- 'anteparo' -> 'screen' ou 'detector screen' (NAO 'shield')\n"
    f"- 'comprimento de onda' -> 'wavelength' (termo unico em ingles)\n"
    f"Escreva as explicacoes em portugues sem acentos.",
    system="Voce e um tradutor tecnico especializado em Fisica. "
           "Mantenha terminologia cientifica padrao em ingles. "
           "Explicacoes em portugues sem acentos."
)
print(traducao)
```

```python
# Traducao inversa: ingles -> portugues tecnico
texto_en = """
The experimental setup employed a Michelson interferometer to measure
the refractive index of a thin glass plate. A coherent light source
(He-Ne laser, 632.8 nm) was split into two beams by a 50/50 beam
splitter. The sample was placed in one arm of the interferometer, and
the resulting fringe shift was recorded using a CCD camera. The
refractive index was determined from the fringe displacement using
the relation n = 1 + N*lambda/(2*d), where N is the number of
fringes shifted and d is the plate thickness.
"""

traducao_inv = ask(
    f"Traduza o seguinte texto tecnico de Fisica do ingles para o portugues "
    f"(sem acentos). Mantenha os termos tecnicos corretos.\n\n"
    f"TEXTO ORIGINAL:\n{texto_en}\n\n"
    f"Forneca a traducao e uma lista de termos tecnicos com a traducao adotada.",
    system="Voce e um tradutor tecnico de Fisica. Traduza para portugues sem acentos. "
           "Use terminologia padrao em portugues."
)
print(traducao_inv)
```

## Formatacao LaTeX com Claude

LaTeX e o padrao para documentos academicos em Fisica. O Claude pode converter
texto e equacoes para LaTeX, gerar tabelas formatadas e criar entradas BibTeX.

```python
# Converter texto com equacoes para LaTeX
texto_para_latex = """
A condicao para minimos de difracao em uma fenda simples e dada por
a * sen(theta) = m * lambda, onde a e a largura da fenda, theta e o
angulo de difracao, m e a ordem do minimo (m = +/-1, +/-2, ...) e
lambda e o comprimento de onda. Para angulos pequenos, podemos usar
a aproximacao sen(theta) ~ tan(theta) = y/L, onde y e a posicao do
minimo no anteparo e L e a distancia fenda-anteparo. Assim, a largura
da fenda pode ser calculada como a = m * lambda * L / y_m.

A incerteza na largura da fenda e obtida por propagacao de incertezas:
sigma_a = a * sqrt( (sigma_y/y)^2 + (sigma_L/L)^2 )
"""

latex_output = ask(
    f"Converta o seguinte texto para formato LaTeX completo, pronto para compilar. "
    f"Inclua:\n"
    f"- Equacoes numeradas usando o ambiente equation\n"
    f"- Simbolos gregos com comandos LaTeX corretos\n"
    f"- Labels para referencia cruzada\n\n"
    f"TEXTO:\n{texto_para_latex}\n\n"
    f"Gere APENAS o trecho LaTeX (nao o preambulo completo do documento).",
    system="Voce e um especialista em LaTeX para documentos de Fisica."
)
print(latex_output)
```

```python
# Gerar uma tabela LaTeX a partir de dados
tabela_latex = ask(
    "Gere uma tabela em LaTeX (ambiente table com tabular) com os seguintes dados "
    "de difracao de fenda simples:\n\n"
    "Ordem m | Posicao (+) mm | Posicao (-) mm | Media mm | a calculado (mm)\n"
    "1       | 9.6            | 9.4            | 9.5      | 0.0999\n"
    "2       | 19.1           | 19.3           | 19.2     | 0.0988\n"
    "3       | 28.5           | 28.8           | 28.65    | 0.0993\n\n"
    "A tabela deve ter:\n"
    "- Caption descritivo\n"
    "- Label para referencia\n"
    "- Linhas horizontais com booktabs (toprule, midrule, bottomrule)\n"
    "- Alinhamento adequado das colunas\n"
    "- Nota de rodape com a incerteza",
    system="Voce e um especialista em tabelas LaTeX com booktabs. "
           "Use o pacote siunitx para numeros quando apropriado."
)
print(tabela_latex)
```

```python
# Gerar entradas BibTeX
bibtex = ask(
    "Gere entradas BibTeX para as seguintes referencias (invente dados realistas):\n\n"
    "1. Um livro classico de optica (ex: Hecht, Optics)\n"
    "2. Um artigo sobre difracao em Physical Review\n"
    "3. Uma apostila de laboratorio de Fisica da UFRGS\n"
    "4. Um livro de Fisica experimental (ex: Taylor, An Introduction to Error Analysis)\n\n"
    "Use chaves BibTeX descritivas (ex: hecht2017optics). "
    "Inclua todos os campos obrigatorios para cada tipo de entrada.",
    system="Voce e um bibliotecario especializado em referencias de Fisica. "
           "Use o formato BibTeX padrao."
)
print(bibtex)
```
