```python
import os
from dotenv import load_dotenv
import anthropic

load_dotenv()
client = anthropic.Anthropic(api_key=os.getenv('ANTHROPIC_API_KEY'))
HAIKU  = 'claude-haiku-4-5-20251001'
SONNET = 'claude-sonnet-4-5'

def ask(prompt, system=None, model=HAIKU, max_tokens=1024):
    kw = dict(model=model, max_tokens=max_tokens,
              messages=[{'role':'user','content':prompt}])
    if system: kw['system'] = system
    return client.messages.create(**kw).content[0].text

print('Pronto!')
```

```python
SYSTEM_RH = """
Voce e o coordenador de gestao de pessoas da Enfitec, empresa junior
de Engenharia Fisica da UFRGS. Suas responsabilidades:

- Desenvolver processos seletivos justos e eficazes
- Criar cases tecnicos que avaliem competencias relevantes
- Elaborar rubricas de avaliacao objetivas
- Fornecer feedback construtivo a candidatos
- Garantir que o processo atraia talentos de fisica e engenharia
- Valorizar diversidade de habilidades: hardware, software, gestao, comunicacao

O perfil desejado: estudantes de Engenharia Fisica ou areas afins da UFRGS,
com interesse em projetos praticos e empreendedorismo.

Sempre responda em portugues brasileiro.
"""

print('System prompt de RH carregado.')
```

## Gerando Cases Tecnicos

Cases tecnicos sao a melhor forma de avaliar competencias praticas.
Vamos gerar um case de hardware e um de software.

```python
prompt_case_hw = """
Crie um case tecnico de HARDWARE para o processo seletivo da Enfitec.

Tema: Projetar um circuito de monitoramento de temperatura

O case deve conter:
1. **Contexto**: Cenario realista de um cliente
2. **Requisitos tecnicos**:
   - Faixa de temperatura a medir
   - Precisao necessaria
   - Interface de saida (digital/analogica)
   - Alimentacao disponivel
3. **Tarefas para o candidato**:
   - Escolher o sensor adequado (justificar)
   - Desenhar diagrama de blocos do circuito
   - Especificar componentes principais
   - Estimar custo do BOM
4. **Criterios de avaliacao** (o que voce espera ver)
5. **Tempo sugerido**: 2-3 horas
6. **Material de apoio**: o que o candidato pode consultar

Nivel: iniciante a intermediario (estudantes do 3o-6o semestre).
"""

case_hw = ask(prompt_case_hw, system=SYSTEM_RH, model=SONNET, max_tokens=2048)
print("=== CASE DE HARDWARE ===")
print(case_hw)
```

```python
prompt_case_sw = """
Crie um case tecnico de SOFTWARE para o processo seletivo da Enfitec.

Tema: Construir um pipeline de dados para sensores

O case deve conter:
1. **Contexto**: Sistema IoT que coleta dados de 10 sensores de temperatura
   a cada 5 segundos
2. **Requisitos tecnicos**:
   - Ler dados de um arquivo CSV simulado
   - Limpar dados (remover outliers, preencher gaps)
   - Calcular estatisticas (media, desvio padrao, tendencia)
   - Gerar alerta se temperatura sair da faixa
   - Visualizar com grafico
3. **Tarefas para o candidato**:
   - Escrever script Python funcional
   - Documentar decisoes de design
   - Explicar como escalaria para 1000 sensores
4. **Criterios de avaliacao**
5. **Tempo sugerido**: 3-4 horas
6. **Dados de teste**: fornecer exemplo do CSV

Nivel: iniciante a intermediario.
"""

case_sw = ask(prompt_case_sw, system=SYSTEM_RH, model=SONNET, max_tokens=2048)
print("=== CASE DE SOFTWARE ===")
print(case_sw)
```

## Rubricas de Avaliacao

Rubricas garantem avaliacao objetiva e consistente entre diferentes
avaliadores. Cada criterio tem escala e descritores claros.

```python
prompt_rubrica = """
Gere rubricas de avaliacao detalhadas para o processo seletivo da Enfitec.

Crie 4 rubricas, uma para cada competencia:

1. **Conhecimento Tecnico** (peso 30%)
2. **Resolucao de Problemas** (peso 30%)
3. **Comunicacao** (peso 20%)
4. **Trabalho em Equipe** (peso 20%)

Para cada rubrica:
- Escala de 1 a 5
- Descritores claros para cada nivel (o que o candidato demonstra)
- Exemplos concretos de comportamentos/respostas em cada nivel
- Formato: tabela Markdown

A rubrica deve ser aplicavel tanto ao case tecnico quanto a entrevista.
"""

rubricas = ask(prompt_rubrica, system=SYSTEM_RH, model=SONNET, max_tokens=3000)
print(rubricas)
```

## Perguntas para Entrevista Tecnica

Perguntas estruturadas por area de conhecimento, com respostas
esperadas e perguntas de aprofundamento.

```python
areas = [
    ("Eletronica", "circuitos, sensores, microcontroladores, PCB"),
    ("Programacao", "Python, C/C++, estruturas de dados, algoritmos"),
    ("Materiais e Instrumentacao", "metrologia, caracterizacao, incertezas"),
    ("Gestao de Projetos", "planejamento, prazos, comunicacao, metodologias ageis"),
]

for area, temas in areas:
    prompt = f"""
Gere 5 perguntas de entrevista tecnica para a area de {area}.
Temas relevantes: {temas}

Para cada pergunta:
- A pergunta principal
- Resposta esperada (resumo, 2-3 linhas)
- Uma pergunta de aprofundamento (follow-up)
- Nivel de dificuldade: Basico / Intermediario / Avancado

As perguntas devem ser adequadas para estudantes de graduacao
em Engenharia Fisica (3o-8o semestre).
"""
    resultado = ask(prompt, system=SYSTEM_RH, model=SONNET, max_tokens=1500)
    print(f"\n{'='*60}")
    print(f"AREA: {area.upper()}")
    print(f"{'='*60}")
    print(resultado)
```

## Avaliando Respostas de Candidatos

Usando Claude para avaliar objetivamente respostas de candidatos
contra a rubrica estabelecida. Demonstracao com respostas de
diferentes niveis de qualidade.

```python
# Simulando respostas de candidatos ao case de hardware
resposta_excelente = """
Para o monitoramento de temperatura, escolhi o sensor DS18B20 (digital,
protocolo 1-Wire) por sua precisao de +/- 0.5C na faixa de -10C a 85C,
que atende aos requisitos. A interface digital simplifica o circuito e
permite conectar multiplos sensores em barramento unico.

Diagrama de blocos: Sensor DS18B20 -> Barramento 1-Wire -> ESP32 ->
WiFi -> Servidor MQTT -> Dashboard web.

Componentes: DS18B20 (R$ 15), ESP32 (R$ 55), resistor 4.7k pull-up,
fonte 5V (R$ 25), PCB prototipo (R$ 10). Total por no: ~R$ 105.

Para escalar, usaria barramento CAN para longas distancias e
adicionaria redundancia com sensor analogico NTC como backup.
"""

resposta_fraca = """
Eu usaria um termometro digital comum ligado no Arduino.
O Arduino manda os dados pro computador pela USB.
Nao sei quanto custa mas acho que da uns 50 reais.
"""

for nome, resposta in [("Candidato A (forte)", resposta_excelente),
                        ("Candidato B (fraco)", resposta_fraca)]:
    prompt_avaliacao = f"""
Avalie a seguinte resposta ao case tecnico de hardware usando a rubrica.

RESPOSTA DO CANDIDATO:
{resposta}

Avalie nos criterios:
1. Conhecimento Tecnico (1-5): escolha do sensor, entendimento do circuito
2. Resolucao de Problemas (1-5): abordagem, justificativas, criatividade
3. Comunicacao (1-5): clareza, organizacao, termos tecnicos corretos

Para cada criterio: nota, justificativa e evidencia na resposta.
Ao final: nota geral ponderada e recomendacao (aprovar/reprovar/reavaliar).
"""
    avaliacao = ask(prompt_avaliacao, system=SYSTEM_RH, model=SONNET, max_tokens=1500)
    print(f"\n{'='*60}")
    print(f"AVALIACAO: {nome}")
    print(f"{'='*60}")
    print(avaliacao)
```

## Feedback Estruturado

Feedback construtivo e importante tanto para candidatos aprovados
quanto reprovados. Vamos gerar feedback personalizado.

```python
cenarios_feedback = [
    {
        "nome": "Ana",
        "resultado": "aprovada",
        "notas": {"tecnico": 4, "problemas": 5, "comunicacao": 4, "equipe": 3},
        "destaques": "Excelente raciocinio tecnico, solucao criativa no case",
        "melhorias": "Pode melhorar trabalho em equipe, um pouco individualista na dinamica"
    },
    {
        "nome": "Bruno",
        "resultado": "reprovado",
        "notas": {"tecnico": 2, "problemas": 3, "comunicacao": 4, "equipe": 5},
        "destaques": "Otimo comunicador, liderou bem a dinamica de grupo",
        "melhorias": "Conhecimento tecnico ainda basico, precisa estudar mais eletronica"
    }
]

for candidato in cenarios_feedback:
    prompt_feedback = f"""
Gere um email de feedback para o seguinte candidato:

Nome: {candidato['nome']}
Resultado: {candidato['resultado']}
Notas: Tecnico={candidato['notas']['tecnico']}/5,
       Resolucao de Problemas={candidato['notas']['problemas']}/5,
       Comunicacao={candidato['notas']['comunicacao']}/5,
       Trabalho em Equipe={candidato['notas']['equipe']}/5
Destaques: {candidato['destaques']}
Pontos de melhoria: {candidato['melhorias']}

O feedback deve:
- Ser empativo e construtivo
- Destacar pontos fortes primeiro
- Apresentar areas de melhoria como oportunidades de desenvolvimento
- Se aprovado: dar boas-vindas e proximos passos
- Se reprovado: encorajar a tentar novamente, sugerir recursos de estudo
- Tom: caloroso, profissional, motivador
"""
    feedback = ask(prompt_feedback, system=SYSTEM_RH, model=SONNET, max_tokens=1200)
    print(f"\n{'='*60}")
    print(f"FEEDBACK: {candidato['nome']} ({candidato['resultado']})")
    print(f"{'='*60}")
    print(feedback)
```

## Planejamento do Processo Seletivo

Cronograma completo do processo seletivo: da abertura das
inscricoes ate a divulgacao dos resultados.

```python
prompt_planejamento = """
Gere um planejamento completo do processo seletivo da Enfitec para o
segundo semestre de 2026.

Parametros:
- Vagas: 8 (4 hardware, 2 software, 2 gestao)
- Candidatos esperados: 30-50
- Periodo: agosto a setembro de 2026
- Equipe organizadora: 3 membros

O planejamento deve conter:

1. **Cronograma detalhado** (tabela com datas, atividade, responsavel)
   - Divulgacao e inscricoes
   - Triagem de curriculos
   - Envio do case tecnico
   - Prazo de entrega do case
   - Avaliacao dos cases
   - Entrevistas individuais
   - Dinamica de grupo
   - Deliberacao final
   - Divulgacao de resultados
   - Periodo de integracao

2. **Canais de divulgacao** (lista com estrategia para cada)

3. **Checklist de materiais necessarios**
   - Formulario de inscricao
   - Cases tecnicos
   - Rubricas de avaliacao
   - Roteiro de entrevista
   - Templates de email (convite, feedback, resultado)

4. **Metricas de sucesso** do processo seletivo

5. **Plano de contingencia** (poucos candidatos, muitas desistencias)

Formato: Markdown estruturado.
"""

planejamento = ask(prompt_planejamento, system=SYSTEM_RH, model=SONNET, max_tokens=3000)
print(planejamento)
```
