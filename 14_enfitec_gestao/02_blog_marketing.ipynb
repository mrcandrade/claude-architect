```python
import os
from dotenv import load_dotenv
import anthropic

load_dotenv()
client = anthropic.Anthropic(api_key=os.getenv('ANTHROPIC_API_KEY'))
HAIKU  = 'claude-haiku-4-5-20251001'
SONNET = 'claude-sonnet-4-5'

def ask(prompt, system=None, model=HAIKU, max_tokens=1024):
    kw = dict(model=model, max_tokens=max_tokens,
              messages=[{'role':'user','content':prompt}])
    if system: kw['system'] = system
    return client.messages.create(**kw).content[0].text

print('Pronto!')
```

```python
SYSTEM_MARKETING = """
Voce e o responsavel de marketing de conteudo da Enfitec, empresa junior
de Engenharia Fisica da UFRGS. Sua missao:

- Criar conteudo que conecta ciencia e engenharia com o publico geral
- Traduzir conceitos complexos de fisica em linguagem acessivel
- Promover a Enfitec como referencia em inovacao e tecnologia
- Manter tom profissional mas envolvente e didatico
- Usar estrategias de SEO e marketing digital
- Conhecer o publico-alvo: estudantes, empresas, comunidade academica

Sempre responda em portugues brasileiro.
"""

print('System prompt de marketing carregado.')
```

## Calendario Editorial com Claude

Um calendario editorial ajuda a manter consistencia na producao de conteudo.
Vamos pedir ao Claude para gerar um plano de 3 meses.

```python
prompt_calendario = """
Gere um calendario editorial de 3 meses (12 semanas) para o blog da Enfitec.

Requisitos:
- 1 post por semana
- Alternar entre categorias:
  * Fisica explicada (conceitos acessiveis)
  * Engenharia na pratica (projetos e aplicacoes)
  * Inovacao e tecnologia (tendencias do mercado)
  * Vida na EJ (bastidores, equipe, eventos)
- Para cada post, inclua: semana, titulo, categoria, resumo (1 linha),
  palavras-chave sugeridas
- Formato: tabela Markdown

Considere datas comemorativas relevantes: Dia do Engenheiro, semanas
academicas, feiras de tecnologia.
"""

calendario = ask(prompt_calendario, system=SYSTEM_MARKETING, model=SONNET, max_tokens=2048)
print(calendario)
```

## Artigo Tecnico-Divulgativo

Conteudo que explica conceitos de fisica de forma acessivel, mas
mantendo rigor tecnico. Ideal para atrair trafego organico e
demonstrar expertise.

```python
prompt_artigo = """
Escreva um artigo de blog com o titulo:
"Como Funciona um Sensor Piezoeletrico? Do Cristal ao Dado Digital"

Requisitos:
- Publico: estudantes de engenharia e gestores de industria
- Extensao: 600-800 palavras
- Estrutura:
  1. Introducao com gancho (por que isso importa?)
  2. O efeito piezoeletrico explicado de forma simples
  3. Como o sensor converte vibracao em sinal eletrico
  4. Aplicacoes industriais reais
  5. Como a Enfitec utiliza essa tecnologia
  6. Conclusao com CTA (call to action)
- Tom: didatico, envolvente, acessivel
- Incluir sugestoes de onde colocar imagens/diagramas [IMAGEM: descricao]
- Formato Markdown
"""

artigo = ask(prompt_artigo, system=SYSTEM_MARKETING, model=SONNET, max_tokens=2048)
print(artigo)
```

## Posts para Redes Sociais

Adaptando conteudo do blog para diferentes plataformas,
cada uma com seu tom e formato.

```python
prompt_social = """
A partir do tema "sensores piezoeletricos na industria", gere posts
para as seguintes plataformas:

1. **LinkedIn** (2 posts):
   - Tom: profissional, tecnico, thought leadership
   - Formato: texto longo (ate 200 palavras), com hashtags relevantes
   - Foco: valor para a industria, cases de sucesso

2. **Instagram** (2 posts):
   - Tom: visual, didatico, acessivel
   - Formato: legenda curta (ate 100 palavras), emojis moderados, hashtags
   - Sugestao de imagem para cada post
   - Foco: curiosidade cientifica, bastidores

3. **Twitter/X** (3 tweets):
   - Tom: conciso, informativo, com gancho
   - Formato: ate 280 caracteres cada
   - Foco: fatos rapidos, links para blog

Para cada post, indique o melhor horario/dia para publicacao.
"""

posts_social = ask(prompt_social, system=SYSTEM_MARKETING, model=SONNET, max_tokens=2048)
print(posts_social)
```

## SEO e Palavras-chave

Otimizacao para mecanismos de busca: palavras-chave, meta descricoes
e variacoes de titulo para melhorar o alcance organico.

```python
prompt_seo = """
Para um artigo de blog sobre "sensores piezoeletricos na industria",
gere uma estrategia de SEO completa:

1. **Palavra-chave principal** e justificativa
2. **Palavras-chave secundarias** (5-8 termos)
3. **Palavras-chave de cauda longa** (5 frases)
4. **Variacoes de titulo** (5 opcoes otimizadas para SEO)
5. **Meta descricao** (ate 160 caracteres, 3 variacoes)
6. **Estrutura de headings** (H1, H2, H3) otimizada
7. **Sugestoes de links internos** (para outros possiveis artigos)
8. **Texto alternativo** para 3 imagens sugeridas

Considere o publico brasileiro e termos de busca em portugues.
"""

seo = ask(prompt_seo, system=SYSTEM_MARKETING, model=SONNET, max_tokens=1500)
print(seo)
```

## Case Studies de Projetos

Cases de sucesso sao poderosos para marketing B2B. Vamos
transformar descricoes de projetos em narrativas convincentes.

```python
descricao_projeto = """
Projeto: Sistema de monitoramento de temperatura em camaras frigorificas
Cliente: Distribuidora de alimentos (medio porte, RS)
Problema: Perdas de R$ 50.000/mes por falhas no controle de temperatura
Solucao: Rede de sensores IoT com alertas em tempo real via app
Resultado: Reducao de 95% nas perdas, ROI em 2 meses
Duracao: 6 semanas
Equipe: 4 membros Enfitec
"""

prompt_case = f"""
Transforme a descricao abaixo em um case study profissional para o blog.

DESCRICAO DO PROJETO:
{descricao_projeto}

Estrutura do case study:
1. **Titulo chamativo** (com resultado mensuravel)
2. **Resumo executivo** (3-4 linhas)
3. **O Desafio** - contexto e problema do cliente
4. **A Solucao** - abordagem tecnica da Enfitec
5. **Os Resultados** - metricas e impacto
6. **Depoimento do cliente** (template para preenchimento)
7. **Proximos passos** - expansao ou novos projetos

Tom: profissional, orientado a resultados, com dados concretos.
Extensao: 400-500 palavras.
"""

case_study = ask(prompt_case, system=SYSTEM_MARKETING, model=SONNET, max_tokens=1500)
print(case_study)
```

## Newsletter para Stakeholders

Uma newsletter mensal mantendo stakeholders informados sobre
atividades, projetos e conquistas da Enfitec.

```python
dados_newsletter = """
Periodo: Marco 2026
Projetos em andamento: 3 (monitoramento industrial, automacao agricola, app mobile)
Projetos concluidos: 1 (sistema de camaras frigorificas)
Novos membros: 5 aprovados no processo seletivo
Eventos: Participacao na Semana de Engenharia UFRGS
Capacitacoes: Workshop de IoT com Professor Dr. Silva
Faturamento do mes: R$ 18.500
Meta trimestral: R$ 50.000 (atingido 62%)
"""

prompt_newsletter = f"""
Redija uma newsletter mensal da Enfitec com base nos dados abaixo.

DADOS:
{dados_newsletter}

Estrutura:
1. Saudacao e destaque do mes (1 paragrafo)
2. Projetos: status e conquistas
3. Equipe: novos membros e capacitacoes
4. Eventos e participacoes
5. Numeros do mes (faturamento, metas)
6. Previa do proximo mes
7. Encerramento motivacional

Tom: informativo, entusiasmado mas profissional.
Publico: membros, ex-membros, professores orientadores, parceiros.
Formato: Markdown com secoes claras.
Extensao: 400-500 palavras.
"""

newsletter = ask(prompt_newsletter, system=SYSTEM_MARKETING, model=SONNET, max_tokens=1500)
print(newsletter)
```
