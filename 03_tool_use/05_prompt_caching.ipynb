# 05 - Prompt Caching
> Reduzindo custos com cache de system prompts longos

**Modulo:** `03_tool_use` | **Feito com:** [Claude](https://claude.ai) (Anthropic)

---


```python
import os
from dotenv import load_dotenv
import anthropic

load_dotenv()
client = anthropic.Anthropic(api_key=os.getenv('ANTHROPIC_API_KEY'))
HAIKU  = 'claude-haiku-4-5-20251001'
SONNET = 'claude-sonnet-4-5'

def ask(prompt, system=None, model=HAIKU, max_tokens=1024):
    kw = dict(model=model, max_tokens=max_tokens,
              messages=[{'role':'user','content':prompt}])
    if system: kw['system'] = system
    return client.messages.create(**kw).content[0].text

print('Pronto!')
```

```python
import time

BASE = ('Esta e a documentacao tecnica do sistema. ' * 300)

def perguntar(q):
    t0=time.time()
    r=client.messages.create(model=HAIKU,max_tokens=256,
        system=[{'type':'text','text':f'Responda com base nesta doc:\n{BASE}',
                  'cache_control':{'type':'ephemeral'}}],
        messages=[{'role':'user','content':q}])
    return {'texto':r.content[0].text,'tempo':round(time.time()-t0,2),
            'cache_criado':getattr(r.usage,'cache_creation_input_tokens',0),
            'cache_lido':getattr(r.usage,'cache_read_input_tokens',0)}

for i,q in enumerate(['Como instalar?','Como configurar?','Como fazer backup?'],1):
    res=perguntar(q)
    print(f'Q{i}: {res["tempo"]}s | criado={res["cache_criado"]} lido={res["cache_lido"]}')
```
