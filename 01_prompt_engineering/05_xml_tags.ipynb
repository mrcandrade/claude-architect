```python
import os
from dotenv import load_dotenv
import anthropic

load_dotenv()
client = anthropic.Anthropic(api_key=os.getenv('ANTHROPIC_API_KEY'))
HAIKU  = 'claude-haiku-4-5-20251001'
SONNET = 'claude-sonnet-4-5'

def ask(prompt, system=None, model=HAIKU, max_tokens=1024):
    kw = dict(model=model, max_tokens=max_tokens,
              messages=[{'role':'user','content':prompt}])
    if system: kw['system'] = system
    return client.messages.create(**kw).content[0].text

print('Pronto!')
```

```python
import json

doc = 'Receita Q3: R$2.3M (+15%). Novos clientes: 234. Churn: 3.2%.'

prompt = f'''
<document>{doc}</document>

<instructions>
Com base no documento, responda:
1. A empresa esta crescendo? (sim/nao + motivo)
2. Qual o maior risco?
Responda com JSON: {{"crescendo": bool, "motivo": str, "risco": str}}
</instructions>
'''

r = json.loads(ask(prompt))
print(json.dumps(r, indent=2, ensure_ascii=False))
```
