```python
import os
from dotenv import load_dotenv
import anthropic

load_dotenv()
client = anthropic.Anthropic(api_key=os.getenv('ANTHROPIC_API_KEY'))
HAIKU  = 'claude-haiku-4-5-20251001'
SONNET = 'claude-sonnet-4-5'

def ask(prompt, system=None, model=HAIKU, max_tokens=1024):
    kw = dict(model=model, max_tokens=max_tokens,
              messages=[{'role':'user','content':prompt}])
    if system: kw['system'] = system
    return client.messages.create(**kw).content[0].text

print('Pronto!')
```

```python
sys_analista = '''
Voce e um analista de dados senior especializado em Python e SQL.
- Sempre mostre codigo executavel
- Explique decisoes tecnicas brevemente
- Responda em portugues brasileiro
'''

for p in ['Como calcular a mediana no pandas?', 'Qual o sentido da vida?']:
    print(f'>>> {p}')
    print(ask(p, system=sys_analista)[:200])
    print('-'*40)
```

```python
import json

sys_json = '''
Voce e um classificador de sentimentos.
Responda SEMPRE com JSON: {"sentimento": "positivo|negativo|neutro", "confianca": 0.0-1.0}
Sem texto adicional.
'''

frases = ['Adorei o produto!', 'Pessimo, nao recomendo.', 'Chegou no prazo.']
for f in frases:
    r = json.loads(ask(f, system=sys_json))
    print(f"{r['sentimento']:10} [{r['confianca']:.0%}] {f}")
```
