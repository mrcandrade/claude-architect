```python
import os
from dotenv import load_dotenv
import anthropic

load_dotenv()
client = anthropic.Anthropic(api_key=os.getenv('ANTHROPIC_API_KEY'))
HAIKU  = 'claude-haiku-4-5-20251001'
SONNET = 'claude-sonnet-4-5'

def ask(prompt, system=None, model=HAIKU, max_tokens=1024):
    kw = dict(model=model, max_tokens=max_tokens,
              messages=[{'role':'user','content':prompt}])
    if system: kw['system'] = system
    return client.messages.create(**kw).content[0].text

print('Pronto!')
```

## Os 3 roles: system, user, assistant


```python
# Sem system prompt
r = client.messages.create(model=HAIKU, max_tokens=80,
    messages=[{'role':'user','content':'Quem e voce em uma frase?'}])
print('Sem system:', r.content[0].text)

# Com system prompt
r2 = client.messages.create(model=HAIKU, max_tokens=80,
    system='Voce e um tecnico de TI senior. Seja direto e tecnico.',
    messages=[{'role':'user','content':'Quem e voce em uma frase?'}])
print('Com system:', r2.content[0].text)
```

## Os 4 ingredientes de um bom prompt
1. TAREFA     - O que fazer
2. CONTEXTO   - Informacoes relevantes
3. FORMATO    - Como responder
4. RESTRICOES - O que evitar


```python
fraco = 'Me fale sobre APIs.'

forte = '''
Tarefa: Compare REST vs GraphQL para um dev backend.
Contexto: Ele usa Django e esta escolhendo para um e-commerce.
Formato: Tabela com criterios: performance, flexibilidade, curva de aprendizado.
Restricoes: Seja objetivo. Termine com uma recomendacao clara.
'''

print('--- FRACO ---'); print(ask(fraco))
print('\n--- FORTE ---'); print(ask(forte))
```

## Inspecionando o objeto de resposta


```python
r = client.messages.create(model=HAIKU, max_tokens=64,
    messages=[{'role':'user','content':'Diga ola em 3 idiomas.'}])
print(f'Texto:     {r.content[0].text}')
print(f'Modelo:    {r.model}')
print(f'Stop:      {r.stop_reason}')
print(f'Tokens in: {r.usage.input_tokens}')
print(f'Tokens out:{r.usage.output_tokens}')
custo = r.usage.input_tokens*0.00000025 + r.usage.output_tokens*0.00000125
print(f'Custo:     ~${custo:.6f}')
```
