```python
import os
from dotenv import load_dotenv
import anthropic

load_dotenv()
client = anthropic.Anthropic(api_key=os.getenv('ANTHROPIC_API_KEY'))
HAIKU  = 'claude-haiku-4-5-20251001'
SONNET = 'claude-sonnet-4-5'

def ask(prompt, system=None, model=HAIKU, max_tokens=1024):
    kw = dict(model=model, max_tokens=max_tokens,
              messages=[{'role':'user','content':prompt}])
    if system: kw['system'] = system
    return client.messages.create(**kw).content[0].text

print('Pronto!')
```

```python
for temp in [0.0, 0.5, 1.0]:
    r = client.messages.create(model=HAIKU, max_tokens=60, temperature=temp,
        messages=[{'role':'user','content':'Descreva o por do sol em uma frase poetica.'}])
    print(f'Temp {temp}: {r.content[0].text.strip()}')
```

```python
from pydantic import BaseModel
from typing import List
import json

class Produto(BaseModel):
    nome: str; preco: float; disponivel: bool

class Lista(BaseModel):
    produtos: List[Produto]

prompt = f'''
Liste 3 produtos eletronicos ficticios.
Responda APENAS com JSON seguindo: {Lista.model_json_schema()}
'''
data = Lista.model_validate_json(ask(prompt))
for p in data.produtos:
    print(f"{'sim' if p.disponivel else 'nao'} {p.nome} R${p.preco:.2f}")
```
