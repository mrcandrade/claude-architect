```python
import os
from dotenv import load_dotenv
import anthropic

load_dotenv()
client = anthropic.Anthropic(api_key=os.getenv('ANTHROPIC_API_KEY'))
HAIKU  = 'claude-haiku-4-5-20251001'
SONNET = 'claude-sonnet-4-5'

def ask(prompt, system=None, model=HAIKU, max_tokens=1024):
    kw = dict(model=model, max_tokens=max_tokens,
              messages=[{'role':'user','content':prompt}])
    if system: kw['system'] = system
    return client.messages.create(**kw).content[0].text

print('Pronto!')
```

```python
def solicitar_aprovacao(acao, detalhes):
    alto = any(p in acao.lower() for p in ['deletar','enviar','publicar','transferir'])
    if not alto:
        print(f'  [AUTO] {acao}'); return True
    print(f'\nAPROVACAO NECESSARIA: {acao}')
    print(f'   Detalhes: {detalhes}')
    return input('   Aprovar? (s/n): ').strip().lower()=='s'

acoes=[('ler dados','leitura do relatorio mensal'),
       ('deletar registro','remover user-001 da base'),
       ('enviar email','notificar 500 usuarios sobre manutencao')]

for acao,detalhe in acoes:
    aprovado=solicitar_aprovacao(acao,detalhe)
    print(f'   -> {"Executado" if aprovado else "Cancelado"}\n')
```
