# 04 - Multi-Agent Pipeline

```python
import os
from dotenv import load_dotenv
import anthropic

load_dotenv()
client = anthropic.Anthropic(api_key=os.getenv('ANTHROPIC_API_KEY'))
HAIKU  = 'claude-haiku-4-5-20251001'
SONNET = 'claude-sonnet-4-5'

def ask(prompt, system=None, model=HAIKU, max_tokens=1024):
    kw = dict(model=model, max_tokens=max_tokens,
              messages=[{'role':'user','content':prompt}])
    if system: kw['system'] = system
    return client.messages.create(**kw).content[0].text

print('Pronto!')
```

```python
import concurrent.futures, time

ETAPAS=[('Extrator','Extraia os fatos principais em bullet points.'),
        ('Avaliador','Avalie cada fato: [VERIFICADO], [INCERTO] ou [OPINIAO].'),
        ('Sumarizador','Crie um resumo executivo de 2 linhas.')]

noticia='A TechCorp anunciou receita de R$2.8bi (+15% vs Q2). CEO diz que vai dominar o mercado. EBITDA: R$800M.'

# Serie
resultado=noticia
for nome,sys_ in ETAPAS:
    print(f'  > {nome}'); resultado=ask(resultado,system=sys_,model=HAIKU)
print('\nRESULTADO SERIE:'); print(resultado)
```

```python
# Paralelo
ASPECTOS=[('Financeiro','Foque em numeros e crescimento.'),
          ('Riscos','Identifique riscos e pontos de atencao.'),
          ('Estrategia','Avalie o posicionamento estrategico.')]

def analisar(args):
    aspecto,instrucao=args
    return aspecto, ask(noticia, system=instrucao, model=HAIKU)

t0=time.time()
with concurrent.futures.ThreadPoolExecutor(max_workers=3) as ex:
    resultados=list(ex.map(analisar,ASPECTOS))
print(f'Paralelo: {time.time()-t0:.2f}s')
for asp,r in resultados: print(f'\n=== {asp} ===\n{r[:150]}')
```
